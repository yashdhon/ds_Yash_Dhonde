```python

coin_data=data.groupby('Coin')['Execution Price'].mean().reset_index(name="avg execution price")
coin_data=coin_data.sort_values(by='avg execution price',ascending=False)
coin_data=coin_data.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.barplot(x='Coin',y='avg execution price',data=coin_data)
plt.xticks(rotation=90)
plt.title("Coins With Highest Price")
```
